## 3주차 데이터 시각화 - plot() 함수

#### 빅데이터 전공 20155342 지현규

x와 y의 2개 축을 기준으로 좌표를 찍듯이 그리는 컨셉을 가진 함수
대표적으로 산점도를 표현할 때 간단하게 구현이 가능한 함수

#### 1. 좌표찍기

````r
plot(2, 1)
````

plot() 함수는 x,y 인자 2개가 기본<br/>
이러한 특성으로 복수개의 좌표에 점을 찍을 수 있음.

````r
plot(c(1,2), c(1,5))
````

2개의 점이 (1,1), (2,5) 좌표에 찍힘<br/>
이런 특성을 이용해 데이터셋을 plot() 함수의 x,y인자에게 넘김으로써 산점도 표현가능

````r
names(cars)
````

R 내장객체 cars에서 <br/> speed정보를 담은 벡터 ```cars$speed``` 가 x축<br/> dist정보를 담은 백터 ```cars$dist```가 y축으로 입력<br/>
차량의 속도와 거리간의 관계 대략적으로 살필 수 있는 산점도 그림

````r
plot(cars$speed, cars$dist)
````

````r
plot(dist~speed, data=cars) # 위의 코드와 동일
````

#### type인자

````r
plot(c(1,2), c(1,5), type = "l")
````

##### type 속성값
<table>
    <tr><th>값</th><th>설명</th></tr>
    <tr><td>p</td><td>점</td></tr>
    <tr><td>l</td><td>선</td></tr>
    <tr><td>b</td><td>점과 선 둘다 동시에</td></tr>
    <tr><td>o</td><td>점과 선 둘다 동시에(단, 겹침)</td></tr>
    <tr><td>h</td><td>히스토그램</td></tr>
    <tr><td>s</td><td>계단모양(stair steps)</td></tr>
    <tr><td>S</td><td>계단모양(Upper stair steps)</td></tr>
    <tr><td>n</td><td>좌표찍지않음</td></tr>
 </table>

````r
plot(c(1,2), c(1,5), type="n") # 스케치북 준비
````

#### 이외의 다양한 인자들

````r
?par# 그래픽스 파라미터 사용가능한 인자 알고싶을 때 참고
````

````r
plot(dist~speed, data=cars, cex=2) # 디폴트로 cex=1, 즉 2배크게
````

````r
plot(dist~speed, data=cars, cex=.5) # 2배 작게
````

````r
plot(dist~speed, data=cars, fg="blue")
````

````r
plot(c(1,2), c(1,5), type="l", lty="dashed")
````

````r
plot(dist~speed, data=cars, pch=3) # pch에 내가 원하는 점 표시가능 'x'
````

##### plot() 주요인자
<table>
    <tr><th>인자</th><th>설명</th></tr>
    <tr><td>xlab</td><td>x축이름</td></tr>
    <tr><td>ylab</td><td>y축이름</td></tr>
    <tr><td>main</td><td>그래프제목</td></tr>
    <tr><td>pch</td><td>점의종류</td></tr>
    <tr><td>cex</td><td>점의크기</td></tr>
    <tr><td>col</td><td>색깔</td></tr>
    <tr><td>xlim</td><td>x좌표축 값의 최대 범위</td></tr>
    <tr><td>ylim</td><td>y좌표축 값의 최대 범위</td></tr>
    <tr><td>typr</td><td>그래프의 종류</td></tr>
 </table>

````r
plot(cars$speed, cars$dist, main="자동차 속도별 정지 거리", xlab="속도", ylab="정지거리", col="blue")
cor(cars$speed, cars$dist)
````

#### 그래프 겹치기 par()

````r
y = c(1,3,2,4,8,5,7,6,9)
plot(y, main="Simple Time Series")

plot(y, main="Simple Time Series", ylim=c(1,10), type="l") # 제목, y축범위, type 선 설정

par(new = T)
z = c(6,5,8,4,2,8,1,7,10)
plot(z, ylim=c(0,10), type="b", ylab="", col="red") # type 선, 점 선택, 색상 red
````

#### abline()

````r
plot(cars$speed, cars$dist, xlab="속도", ylab="정지거리", main="정지거리 별 속도")
abline(a=-5, b=3.5, col="red")
abline(h=mean(cars$dist), lty=2, col="blue")
abline(v=mean(cars$speed), lty=3, col="green")
# abline(a=60, b=0, col="blue")
# abline(a=-5, b=3.5, col="red") # 절편 -5, 기울기 b인 직선
# abline(h=mean(cars$dist), lty=2, col="blue") # dist평균 수평선
# abline(v=mean(cars$speed), lty=2, col="green") # speed평균 수직선
````
